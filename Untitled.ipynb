```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

# Data Analysis

1. Missing Values

```python
features = df.columns[1:]
features
```

```python
null_features = []
for feature in features:
    if df[feature].isna().sum() > 0:
        null_features.append(feature)
        print("{} has {}% null values".format(feature, (df[feature].isna().sum()/len(df[feature]))*100))
```

```python
for feature in null_features:
    data = df.copy()
    data[feature] = np.where(data[feature].isna(),1,0)
    data.groupby(feature)["SalePrice"].mean().plot.bar()
    plt.title(feature)
    plt.show()
```

2. Numerial Values

```python
numerical_features = [feature for feature in features if df[feature].dtypes != "O"][:-1]
discrete = [feature for feature in numerical_features if len(df[feature].unique()) < 25]
continuous = list(set(numerical_features) - set(discrete))      
print("Total numerical features: ", len(numerical_features))
print("Total discrete features: ", len(discrete))
print("Total continuous features: ", len(continuous))
```

```python
for feature in discrete:
    data = df.copy()
    data.groupby(feature)["SalePrice"].mean().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.show()
```

```python
for feature in continuous:
    data = df.copy()
#     data[feature] = np.log(data[feature])
    data[feature].hist(bins = 25)
    plt.title(feature)
    plt.show()
```

3. Categorical features


```python
categorical_features = [feature for feature in features if df[feature].dtypes == "O"]
print("Total categorical featues are: ", len(categorical_features))
```

```python
for feature in categorical_features:
    print("{} has {} unique categories".format(feature,len(df[feature].unique())))
```

```python
for feature in categorical_features:
    temp = df.groupby(feature).count()
    temp = temp/temp.sum()
    print("{} has {} categories".format(feature, len(temp[temp["Id"] < .01])))
```

4. Outliers

```python
for feature in numerical_features:
    data = df.copy()
    data.boxplot(column = feature)
    plt.show()
```

```python
for feature in continuous:
    data = df.copy()
    plt.scatter(data[feature], data["SalePrice"])
    plt.show()
```

5. Relation in Independent and Dependent Variable

```python
for feature in numerical_features:
    plt.plot(df[feature],df["SalePrice"])
    plt.title(feature)
    plt.show()
```

# Feature Engineering

### 1. Missing Values


```python
for feature in null_features:
    if feature in categorical_features:
        df[feature].fillna("Missing", inplace = True)
    if feature in numerical_features:
        df[feature].fillna(df[feature].mean(), inplace = True)
```

### 2. Temporal data

```python
date_features = [feature for feature in features if "Yr" in feature or "Year" in feature]
date_features
```

```python
for feature in date_features[:-1]:
    df[feature] = df["YrSold"] - df[feature]
```

### 3. With Rare Categories

```python
count = 0
data = df.copy()
for feature in categorical_features:
    temp = df.groupby(feature).count()
    temp = temp/temp.sum()
    temp_df = temp[temp["Id"] > .01].index
    df[feature] = np.where(df[feature].isin(temp_df), df[feature], "rar" )
```

### 4. Gaussian transformation of continuous data

```python
for feature in list(set(numerical_features)-set(date_features)):
    df[feature] = np.where(df[feature] <=0, df[feature], np.log(df[feature]))
```
